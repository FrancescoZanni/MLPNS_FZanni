```python
data = scipy.io.loadmat('Data.mat')
data["lambda"]=[0.2,1,3,10]
data["ris"]=[0.3572,0.8551,1.7763,3.2976]
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,6))
ax.plot(data["x"].T,data["yL02"].T,label="fit $\lambda$=0.2",linestyle='dashed',color='#FF5995',alpha=0.6)
ax.plot((data["a"]**2).T,data["E0L02"].T,label="sim $\lambda$=0.2",color='#FF5995', marker='o',linewidth=0)


ax.plot(data["x"].T,data["yL1"].T,label="fit $\lambda=1$",linestyle='dashed',color='#1E88E5',alpha=0.6)
ax.plot((data["a"]**2).T,data["E0L1"].T,label="sim $\lambda$=0.2",color='#1E88E5', marker='o',linewidth=0)

ax.plot(data["x"].T,data["yL3"].T,label="fit $\lambda=3$",linestyle='dashed',color='#FFC107',alpha=0.6)
ax.plot((data["a"]**2).T,data["E0L3"].T,label="sim $\lambda$=0.2",color='#FFC107', marker='o',linewidth=0)

ax.plot(data["x"].T,data["yL10"].T,label="fit $\lambda=10$",linestyle='dashed',color='#004D40',alpha=0.6)
ax.plot((data["a"]**2).T,data["E0L10"].T,label="sim $\lambda$=0.2",color='#004D40', marker='o',linewidth=0)

ax.set_yticks(data["ris"])



ax.legend()
ax.set_title("Stima del livello fondamentale E0 ")
ax.set_xlabel("$a^2$")
ax.set_ylabel("E0")
ax.set_xlim(0,1.03)
ax.set_ylim(0.2,3.6)
txt="Calcolo dell'energia del libello fondamentale per un sistema soggetto ad un potenziale a doppia buca \n$\lambda(x^2-x_0^2)^2$ al variare del parametro $\lambda$. L'estrapolazione dell'intercetta fornisce la migliore stima"
fig.text(.153,-.05,txt)
```
